```python
import requests

URL = 'http://localhost:5000/'
#URL = 'http://68.183.114.149:5000/'
```

```python
# init
u1_payload = {'deviceUUID': 'user1'}
u2_payload = {'deviceUUID': 'user2'}
r = requests.get(URL + 'init', params = u1_payload)
u1_datetime = r.json()["datetime"]
print(r.status_code)
print(r.json())
print(u1_datetime)

r = requests.get(URL + 'init', params = u2_payload)
u2_datetime = r.json()["datetime"]
print(r.status_code)
print(r.json())
print(u2_datetime)
```

```python
import numpy as np
import scipy.signal


def read_data(filename, deviceUUID, datetime):
    with open(filename, 'r') as f:
        rows = f.read().split('|')[4:]
    
    ret = {
        'flex' : [],
        'ecg' : [],
        'eda' : [],
        'deviceUUID': deviceUUID,
        'datetime': datetime
    }
    for i in range(len(rows)):
        x = [int(val) for val in rows[i].split(',')]
        ret['flex'].append(x[0])
        ret['ecg'].append(x[1])
        ret['eda'].append(x[2])
        
    return ret
```

```python
# upload_train_data
json = read_data('maggieawake', u1_payload['deviceUUID'], u1_datetime)
print(json)
r = requests.post(URL + 'upload', json=json)
print(r.status_code)
print(r.json())

json = read_data('maggieawake2',u2_payload['deviceUUID'], u2_datetime)
print(json)
r = requests.post(URL + 'upload', json=json)
print(r.status_code)
print(r.json())
```

```python
# train
u1_payload["datetime"] = u1_datetime
u2_payload["datetime"] = u2_datetime
r = requests.get(URL + 'train', params = u1_payload)
print(r.status_code)
print(r.json())
r = requests.get(URL + 'train', params = u2_payload)
print(r.status_code)
print(r.json())
```

```python
# upload more data
json = read_data('maggieasleep', u1_payload['deviceUUID'], u1_datetime)
r = requests.post(URL + 'upload', json=json)
print(r.status_code)
print(r.json())

json = read_data('maggieasleep',u2_payload['deviceUUID'], u2_datetime)
r = requests.post(URL + 'upload', json=json)
print(r.status_code)
print(r.json())
```

```python
# predict
r = requests.post(URL + 'predict', params = u1_payload)
print(r.status_code)
print(r.json())

r = requests.post(URL + 'predict', params = u2_payload)
print(r.status_code)
print(r.json())
```
