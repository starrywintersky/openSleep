```python
import requests

URL = 'http://localhost:5000/'
#URL = 'http://68.183.114.149:5000/'
```

```python
# init
payload = {'deviceUUID': 'user2'}
r = requests.get(URL + 'init', params = payload)
datetime = r.json()["datetime"]
print(r.status_code)
print(r.json())
print(datetime)
```

```python
import numpy as np
import scipy.signal


def read_data(filename):
    with open(filename, 'r') as f:
        rows = f.read().split('|')[4:]
    
    ret = {
        'flex' : [],
        'ecg' : [],
        'eda' : [],
        'deviceUUID': payload['deviceUUID'],
        'datetime': datetime
    }
    for i in range(len(rows)):
        x = [int(val) for val in rows[i].split(',')]
        ret['flex'].append(x[0])
        ret['ecg'].append(x[1])
        ret['eda'].append(x[2])
        
    return ret
```

```python
# upload_train_data
json = read_data('maggieawake')
r = requests.post(URL + 'upload', json=json)
print(r.status_code)
print(r.json())
```

```python
# train
payload["datetime"] = datetime
r = requests.get(URL + 'train', params = payload)
print(r.status_code)
print(r.json())
```

```python
# upload more data
json = read_data('maggieasleep')
r = requests.post(URL + 'upload', json=json)
print(r.status_code)
print(r.json())
```

```python
# predict
r = requests.post(URL + 'predict', params = payload)
print(r.status_code)
print(r.json())
```

```python
# get data
r = requests.get(URL + 'data', params = payload)
print(r.status_code)
print(r.text)
```
